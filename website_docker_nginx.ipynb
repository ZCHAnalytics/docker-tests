# Creating a website with Docker and NGINX

## 1. Prepare the environment:
Ensure the Docker Desktop is installed and running on the local computer.

## 2. Organize Website Files: 
Create a folder named 'website' on the local computer. Place all the files required for the website, such as HTML, CSS, and JavaScript files, inside this folder.

## 4. Navigate to Website Folder: 
In PowerShell, change directory (`cd`) to the 'website' folder.

##3. In Windows PowerShell, navigate to the local directory ‘website’.

## 4. Run NGINX Container: 
Execute the docker run command to create and start a new container based on the NGINX image. Use the appropriate options to configure the container:  
- `--rm` to remove the container after it exits, keeping the environment clean
- `--name` to assign a custom name to the container, making it easier to manage
- `-p` to map a port on the local machine (e.g.8080) to the port inside the container (e.g.80), allowing access to the website
- `-v` to mount the 'website' folder to a directory within the container where NGINX expects files (e.g.`/usr/share/nginx/html`)
- end the command line with NGINX to pull an NGINX Docker images with the required configuration files.

## 5. Test the website:
In a web browser, navigate to `http://localhost:8080` to check if the website is accessible and read an inspirational quote from Paulo Coelho. 

## 6. Check Container Status: 
Stop the container (using CTRL + C). To verify that the container was removed automatically after it stopped (as specified by the --rm flag), use the docker ps -a command to list all containers, including those that have stopped.

## 7. Optional: Remove NGINX Image:
If the NGINX image is no longer needed, use the `docker rmi` command to remove it.
